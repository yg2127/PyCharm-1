```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# 데이터 생성 (300개의 샘플, 4개의 클러스터 중심)
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.7, random_state=42)

# 데이터 산점도 시각화
plt.scatter(X[:, 0], X[:, 1], s=30)
plt.title('Sample Data (make_blobs)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

# 병합형 계층 군집화 수행 (Ward 연결법)
Z = linkage(X, method='ward')

# 덴드로그램 시각화
plt.figure(figsize=(10, 5))
dendrogram(Z, truncate_mode='lastp', p=5, leaf_rotation=45, leaf_font_size=10)
plt.title('Hierarchical Clustering Dendrogram (Ward)')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

# Agglomerative Clustering (군집 수 4로 설정)
agg = AgglomerativeClustering(n_clusters=3, linkage='ward')
labels = agg.fit_predict(X)

# 결과 산점도 시각화 (군집별 색상 표시)
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='rainbow', s=30)
plt.title('Agglomerative Clustering Result (Ward)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.show()
```
